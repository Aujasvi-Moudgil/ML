# Modelling using Logistic Regression,KNN,NaiveBayes,DecisionTree,Random Forest


```python
# Independent, dependent variables
X=train.drop(['PassengerId','Survived'],axis=1)
Y=train.Survived
X.columns
```

```python
# train test split 
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

# GRidsearch to find the best Hyper parameters of Decision tree, KNN and  Random Forest

```python
parameter_rf= {'n_estimators':np.arange(1,100),'max_depth':np.arange(1,10)} 
# we are restricting the depth to 10 layers
model=RandomForestClassifier(criterion='entropy')
parameter_dt = {'max_depth':np.arange(1,10)} 
# we are restricting the depth to 10 layers
model1=DecisionTreeClassifier(criterion='entropy')
parameter_kn = {'n_neighbors':np.arange(1,15)} 
# we are restricting the depth to 10 layers
model2=KNeighborsClassifier(weights='uniform',metric='euclidean')
```

```python
GS_dt= GridSearchCV(model1,parameter_dt,cv=5) # we give the model created  above inside the GridSearch to give best number of Parameters
GS_dt.fit(X,Y)
print(GS_dt.best_params_)
```

```python
GS_kn= GridSearchCV(model2,parameter_kn,cv=5) # we give the model created  above inside the GridSearch to give best number of Parameters
GS_kn.fit(X,Y)
print(GS_kn.best_params_)
```

```python
GS_rf= GridSearchCV(model,parameter_rf,cv=5) # we give the model created  above inside the GridSearch to give best number of Parameters
GS_rf.fit(X,Y)
print(GS_rf.best_params_)
```

# All model fitting and Prediction using the training and testing created on the Train dataset

```python
# Model Comparison's for 5 models
# Decision Tree
DT_model=DecisionTreeClassifier(max_depth=4,criterion='entropy')
DT_model.fit(x_train,y_train)
pred1=DT_model.predict(x_test)

print("Confusion matrix for DT:\n",metrics.confusion_matrix(y_test,pred1))
# Random Forest
Rf_model=RandomForestClassifier(n_estimators=54,max_depth=9,criterion='entropy')
Rf_model.fit(x_train,y_train)
pred2=Rf_model.predict(x_test)

print("\nConfusion matrix for Random Forest:\n",metrics.confusion_matrix(y_test,pred2))
#Gaussian Naive Bayes
Nb_model=GaussianNB()
Nb_model.fit(x_train,y_train)
pred3=Nb_model.predict(x_test)


print("\nConfusion matrix for Naive Bayes:\n",metrics.confusion_matrix(y_test,pred3))
# logistic Regression
LOr_model=LogisticRegression()
LOr_model.fit(x_train,y_train)
pred4=LOr_model.predict(x_test)


print("\nConfusion matrix for Logistic Regression:\n",metrics.confusion_matrix(y_test,pred4))
#KNN
Kn_model=KNeighborsClassifier(n_neighbors=3,weights='uniform',metric='euclidean')
Kn_model.fit(x_train,y_train)
pred5=Kn_model.predict(x_test)


print("\nConfusion matrix for Knn:\n",metrics.confusion_matrix(y_test,pred5))
#knn_model==KNeighborsRegressor(n_neighbors=k,weights='uniform',metric='euclidean')
```

```python
# Classification Reports for all 5 models
print("DT Classification Report:\n",metrics.classification_report(y_test,pred1))
print("\nRandom Forest Classification Report:\n",metrics.classification_report(y_test,pred2))
print("\nNaive Bayes Classification Report:\n",metrics.classification_report(y_test,pred3))
print("\nLogistic Regression Classification Report:\n",metrics.classification_report(y_test,pred4))
print("\nKNN Classification Report:\n",metrics.classification_report(y_test,pred5))

```

```python
# kfold cross validation
mod=[]
mod.append(('Decision Tree',DT_model))
mod.append(('Random Forest',Rf_model))
mod.append(('Naive Bayes',Nb_model))
mod.append(('Logistic Regression',LOr_model))
mod.append(('KNN',Kn_model))
result=[]
names=[]
scoring='accuracy'
for name,model in mod:
    kfold=model_selection.KFold(n_splits=7,random_state=2,shuffle=True)# 5 fold cross validation function
    cv_results=model_selection.cross_val_score(model,X,Y,cv=kfold,scoring='accuracy')
    result.append(cv_results)
    names.append(name)
    msg="%s:  Performance Range is  (%1.3f %s %1.3f) with a Varience error of (%f) and mean of %1.3f"%(name,cv_results.min()*100,'-',cv_results.max()*100,cv_results.std(),cv_results.mean()*100)
    print(msg)
# boxplot for the models

fig=plt.figure(figsize=[10,8])
fig.suptitle("Algorithm Comparison")
ax=fig.add_subplot(111)
plt.boxplot(result)
ax.set_xticklabels(names)
plt.show()
```
