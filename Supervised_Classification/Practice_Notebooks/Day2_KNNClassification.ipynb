```python
import numpy as np 
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore
from sklearn.metrics import precision_score
```

```python
bc_df = pd.read_csv("wisc_bc_data.csv")
```

```python
bc_df.head() #Diagnosis Benign, Malignent
```

```python
bc_df['diagnosis']=bc_df.diagnosis.astype('category')
```

```python
bc_df.describe().transpose()
```

```python
bc_df = bc_df.drop(['id'],axis=1)
```

```python
bc_df.head()
```

```python
bc_df.corr()
```

```python
bc_feature_df = bc_df.drop(labels='diagnosis',axis=1) #create a sep dataframe only of the features ie. independent attributues
```

```python
bc_feature_df_z = bc_feature_df.apply(zscore)
#convert all attributes to z scale #all value ranges from -3 to +3 (uniform interval) Distance will not be impacted by units
```

```python
bc_feature_df_z.head()
```

```python
Y = bc_df['diagnosis']
```

```python
X = bc_feature_df_z
```

```python
X = np.array(X)
```

```python
Y = np.array(Y)
```

```python
type(X)
```

```python
type(Y)
```

```python
X.shape
```

```python
Y.shape
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
X_Train.shape
```

```python
Y_Test.shape
```

```python
NNH = KNeighborsClassifier(n_neighbors=21, weights='distance',metric='euclidean')
```

```python
NNH.fit(X_Train,Y_Train)
```

```python
y_pred = NNH.predict(X_Test)
```

```python
y_pred
```

```python
acc = metrics.accuracy_score(Y_Test,y_pred)
```

```python
print(acc)
```

```python
cm = metrics.confusion_matrix(Y_Test,y_pred)
print(cm)
```

```python
pd.DataFrame(Y_Test)[0].value_counts()
```

```python
cr = metrics.classification_report(Y_Test,y_pred)
print(cr) #f1 score is the average of precsion and recall
```
