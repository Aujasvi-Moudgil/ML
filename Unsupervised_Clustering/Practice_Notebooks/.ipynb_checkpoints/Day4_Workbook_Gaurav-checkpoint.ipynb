```python
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
iris = iris.drop('Id', axis = 1)
x= iris.SepalLengthCm
y= iris.SepalWidthCm
```

```python
#Standard Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = iris.drop("Species", axis = 1)
y = iris.Species
```

```python
#Standardizing the scale
x = sc.fit_transform(x)
```

```python
#Apply PCA

from sklearn.decomposition import PCA
pc =PCA(n_components= 2)
Pcomp = pc.fit_transform(x)
```

```python
pc.components_
```

```python
pc.explained_variance_ratio_
```

```python
#PLot this

Principaldf = pd.DataFrame(Pcomp, columns = ['PrinComp1', 'PrinComp2'])
finaldf = pd.concat([Principaldf,y], axis=1)
```

```python
finaldf.head()
```

```python
finaldf.shape
```

```python
#PLot this
plt.figure()
plt.scatter(finaldf.PrinComp1,finaldf.PrinComp2)
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x= finaldf.drop('Species', axis=1)
y= finaldf.Species
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 122)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(x_train, y_train)
```

```python
ypred = logreg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, ypred)
```

```python
house=pd.get_dummies(house)
```

```python
house.drop('Id', axis=1, inplace=True)
```

```python
col = house.columns
```

```python
house=sc.fit_transform(house)
```

```python
house = pd.DataFrame(house, columns=col)
```

```python
house.head()
```

```python
x= house.drop('SalePrice', axis=1)
y= house.SalePrice
```

```python
from sklearn.decomposition import PCA
pc =PCA(n_components= 120)
Pcomp = pc.fit_transform(x)
pc.explained_variance_ratio_
```

```python
variance = 0
for i in range(120):
    variance = variance + pc.explained_variance_ratio_[i]
variance
```

```python
x_train, x_test, y_train, y_test = train_test_split(Pcomp,y,test_size=0.3,random_state=122)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(x_train,y_train)
```

```python
ypred= linreg.predict(x_test)
```

```python
np.sqrt(np.mean((ypred-y_test)**2))
```

```python
from statsmodels.api import add_constant
import statsmodels.formula.api as sm

x2= add_constant(Pcomp)
lm= sm.OLS(y,x2)
ln=lm.fit()
print(ln.summary())
```
