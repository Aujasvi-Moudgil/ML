```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook  
# this will allow the 3d plot to render
from mpl_toolkits.mplot3d import Axes3D
df=pd.read_csv("Iris.csv")
df.drop('Id',axis=1,inplace=True)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=df.iloc[:,0]
y=df.iloc[:,1]
ax.scatter(x,y,c=df.iloc[:,0],marker='o')
ax.set_xlabel("sepallenghtcm")
ax.set_ylabel("sepalwidthcm")
plt.show()
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
features=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm',]
y=df.loc[:, 'Species']
x=df.loc[:,features]


x=sc.fit_transform(x)


from sklearn.decomposition import PCA
pc=PCA(n_components=2)
pcomp=pc.fit_transform(x)




pc.components_
```

```python
pc.get_covariance
```

```python
pc.explained_variance_ratio_
```

```python
# let plot this first we will create a DataFrame of this>>>>>>>>>>
pdf=pd.DataFrame(pcomp,columns=['pc1','pc2'])
fdf=pd.concat([pdf,df[['Species']]],axis=1)
```
