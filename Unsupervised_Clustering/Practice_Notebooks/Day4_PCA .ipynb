```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D


#This will allow the 3D plot to Render
import matplotlib.pyplot as plt
%matplotlib notebook 
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

iris = pd.read_csv("Iris.csv")

iris.drop("Id",axis=1)

x = iris['PetalLengthCm']
y = iris['PetalWidthCm']


ax.scatter(x,y,c=iris['SepalLengthCm'],marker='o')
ax.set_xlabel('PetalLengthCm')
ax.set_ylabel('PetalWidthCm')
plt.show()
```

```python
#Standard Scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

#Seprating the features
x = iris.loc[:,features]
y = iris.loc[:,'Species']
```

```python
x = sc.fit_transform(x)
```

```python
#Apply PCA
from sklearn.decomposition import PCA 
pc = PCA(n_components = 2)
Pcomp = pc.fit_transform(x)
```

```python
pc.components_
```

```python
pc.get_covariance
```

```python
pc.explained_variance_ratio_
```

```python
#Lets Plot this-First we will create a DataFrame of this...
Principaldf = pd.DataFrame(Pcomp,columns=['PrinComp1','PrinComp2'])
finaldf = pd.concat([Principaldf, iris[["Species"]]],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = iris.drop("Species",axis=1)
y = iris.Species
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
pred_lg = lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,pred_lg)*100
```
